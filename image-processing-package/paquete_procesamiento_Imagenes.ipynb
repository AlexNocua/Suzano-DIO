```python
from image_processing.processing.image_processing import apply_blur, convert_to_grayscale, resize_image
from image_processing.utils.image_utils import load_image, show_image
```

```python
imagen = "Foto.jpg"

# Aplicacin de las funcionalidades basicas creadas en el paquete.
imagen = load_image(imagen)
show_image(imagen)
```

```python
# procesamientos  basicos creados
imagen_gray = convert_to_grayscale(imagen)
show_image(imagen_gray, "grises")
img_blur = apply_blur(imagen)
show_image(img_blur, 'efecto blur')
imagen_ajustada = resize_image(imagen, 100,100)
show_image(imagen_ajustada, 'imagen redimencionada')
```
